## ç¬¬ 2 éƒ¨åˆ†ï¼šå·¥ä½œè®°å¿† (WorkingMemory)

å·¥ä½œè®°å¿†æ˜¯æœ€ç®€å•çš„è®°å¿†ç±»å‹ï¼š
- âš¡ çº¯å†…å­˜å­˜å‚¨ï¼Œé€Ÿåº¦å¿«
- â° æœ‰ TTLï¼ˆè¿‡æœŸæ—¶é—´ï¼‰
- ğŸ“Š æœ‰å®¹é‡é™åˆ¶

```python
from hello_agent.hello_agents.memory.types import WorkingMemory

# åˆ›å»ºå·¥ä½œè®°å¿†å®ä¾‹
working_memory = WorkingMemory(
    config=MemoryConfig(
        working_memory_capacity=5,  # æœ€å¤š5æ¡è®°å¿†
        working_memory_ttl_minutes=1  # 1åˆ†é’Ÿè¿‡æœŸ
    ),
    user_id="test_user"
)

print("âœ… å·¥ä½œè®°å¿†åˆ›å»ºæˆåŠŸï¼")
print(f"ç±»å‹: {working_memory.memory_type}")
print(f"å®¹é‡é™åˆ¶: {working_memory.config.working_memory_capacity}")
```

### 2.1 æ·»åŠ è®°å¿†

```python
# æ·»åŠ å‡ æ¡è®°å¿†
memories = [
    "æˆ‘å«å¼ ä¸‰",
    "æˆ‘ä»Šå¹´25å²",
    "æˆ‘æ˜¯ä¸€åç¨‹åºå‘˜",
    "æˆ‘å–œæ¬¢Python",
    "æˆ‘æ­£åœ¨å­¦ä¹ AI Agent"
]

memory_ids = []
for content in memories:
    item = MemoryItem(
        id=working_memory._generate_id(),
        content=content,
        memory_type="working",
        user_id="test_user",
        timestamp=datetime.now(),
        importance=0.5
    )
    mem_id = working_memory.add(item)
    memory_ids.append(mem_id)
    print(f"âœ… å·²æ·»åŠ : {content} (ID: {mem_id[:8]}...)")

# æŸ¥çœ‹ç»Ÿè®¡
stats = working_memory.get_stats()
print(f"\nğŸ“Š å½“å‰ç»Ÿè®¡: {stats}")
```

### 2.2 æ£€ç´¢è®°å¿†

```python
# æœç´¢å…³é”®è¯
query = "Python"
results = working_memory.retrieve(query, limit=3)

print(f"ğŸ” æœç´¢ '{query}' çš„ç»“æœï¼š")
for i, item in enumerate(results, 1):
    print(f"{i}. {item.content}")
    print(f"   é‡è¦æ€§: {item.importance}, æ—¶é—´: {item.timestamp.strftime('%H:%M:%S')}")
```

### 2.3 æµ‹è¯•å®¹é‡é™åˆ¶

```python
# å°è¯•æ·»åŠ ç¬¬6æ¡è®°å¿†ï¼ˆè¶…è¿‡å®¹é‡é™åˆ¶ï¼‰
print(f"å½“å‰è®°å¿†æ•°é‡: {working_memory.get_stats()['count']}")
print(f"å®¹é‡é™åˆ¶: {working_memory.config.working_memory_capacity}")

# æ·»åŠ æ–°è®°å¿†
new_item = MemoryItem(
    id=working_memory._generate_id(),
    content="è¿™æ˜¯ç¬¬6æ¡è®°å¿†ï¼Œä¼šè§¦å‘å®¹é‡æ¸…ç†",
    memory_type="working",
    user_id="test_user",
    timestamp=datetime.now(),
    importance=0.9  # é«˜é‡è¦æ€§
)

working_memory.add(new_item)
print(f"\næ·»åŠ åçš„è®°å¿†æ•°é‡: {working_memory.get_stats()['count']}")
print("ğŸ’¡ æ³¨æ„ï¼šå·¥ä½œè®°å¿†ä¼šè‡ªåŠ¨æ¸…ç†ä½é‡è¦æ€§çš„è®°å¿†ä»¥ä¿æŒå®¹é‡é™åˆ¶")
```

MemoryManager ç»Ÿä¸€ç®¡ç†æ‰€æœ‰ç±»å‹çš„è®°å¿†ï¼Œæä¾›ï¼š

```python
# æ·»åŠ å·¥ä½œè®°å¿†ï¼ˆçŸ­æœŸä¿¡æ¯ï¼‰
mem_id_1 = manager.add_memory(
    content="å½“å‰å¯¹è¯çš„ä¸»é¢˜æ˜¯å­¦ä¹ è®°å¿†ç³»ç»Ÿ",
    memory_type="working",
    importance=0.6
)
print(f"âœ… æ·»åŠ å·¥ä½œè®°å¿†: {mem_id_1[:8]}...")

# æ·»åŠ æƒ…æ™¯è®°å¿†ï¼ˆäº‹ä»¶ï¼‰
mem_id_2 = manager.add_memory(
    content="2026å¹´1æœˆ26æ—¥ï¼Œæˆ‘å¼€å§‹å­¦ä¹ HelloAgentsçš„è®°å¿†ç³»ç»Ÿ",
    memory_type="episodic",
    importance=0.8,
    metadata={"event_type": "learning_start", "date": "2026-01-26"}
)
print(f"âœ… æ·»åŠ æƒ…æ™¯è®°å¿†: {mem_id_2[:8]}...")

# æŸ¥çœ‹ç»Ÿè®¡
stats = manager.get_stats()
print(f"\nğŸ“Š è®°å¿†ç»Ÿè®¡:")
for mem_type, count in stats.items():
    print(f"  {mem_type}: {count} æ¡")
```
